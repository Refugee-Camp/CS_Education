```python
%matplotlib inline
from matplotlib.pyplot import imshow
from PIL import Image
import numpy as np
import skimage.color as sc

i = np.array(Image.open('uYbLDJoRkP4.jpg'))
imshow(i)
```

```python
i_mono = sc.rgb2gray(i)
imshow(i_mono,cmap='gray')
i_mono.shape
```

# View pixel value distribution

```python
def im_hist(img):
    import matplotlib.pyplot as plt
    fig = plt.figure(figsize=(8,6))
    fig.clf()
    ax = fig.gca()
    ax.hist(img.flatten(), bins = 256)
    plt.show()
    
im_hist(i_mono)
```

```python
def im_cdf(img):
    import matplotlib.pyplot as plt
    fig = plt.figure(figsize=(8,6))
    fig.clf()
    ax = fig.gca()
    ax.hist(img.flatten(), bins = 256, cumulative = True)
    plt.show()
    
im_cdf(i_mono)
```

# Equalize the image

```python
from skimage import exposure

i_eq = exposure.equalize_hist(i_mono)
imshow(i_eq, cmap='gray')
```

```python
im_hist(i_eq)
im_cdf(i_eq)
```
